## 4.1
**Description:**  
Choose a number between 1 and 10 and assign it to the variable `secret`.  
Then choose another number between 1 and 10 and assign it to the variable `guess`.  
Use if/elif/else statements to print whether the guess is too low, too high, or just right.


```python
# Assignment 4.1
# Compares a guess to a secret number and prints whether it's too low, too high, or just right.

secret = 7
guess = 9

if guess < secret:
    print("Too low")
elif guess > secret:
    print("Too high")
else:
    print("Just right")
```

## 4.2
**Description:**  
Assign True or False to the variables `small` and `green`.  
Then use if/else statements to decide which object matches those traits:  
cherry, pea, watermelon, or pumpkin.


```python
# Assignment 4.2
# Uses boolean logic to print which object matches the size and color.

small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

## 6.2
**Description:**  
Assign the value `7` to the variable `guess_me`, and `1` to the variable `number`.  
Write a `while` loop that compares `number` with `guess_me` and prints:
- "too low" if `number` is less than `guess_me`
- "found it!" if equal (then exits the loop)
- "oops" if greater (then exits the loop)  
Increment `number` at the end of the loop.


```python
# 6.2
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
    number += 1
```

## 6.3
**Description:**  
Assign `5` to `guess_me`.  
Use a `for` loop to iterate `number` over `range(10)`.  
Print "too low" if `number` is less than `guess_me`,  
"found it!" if equal (then break),  
and "oops" if greater (then exit the loop).


```python
# 6.3
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
